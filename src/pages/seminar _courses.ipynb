```python
import pandas as pd
df=pd.read_csv('combined_grad_stats_final.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.columns
```

```python
df.iloc[4651]['Total'] = df.iloc[4651]['Total'] - df.iloc[4651]['Unnamed: 22']
# dropping all the useless columns
df.drop(columns=['Section', 'Unnamed: 22', 'Section', 'II', 'AU', 'PP', 'NP', 'pp', 'W', 'S', 'section_tag_div'], inplace=True)
```

```python
Seminar_courses = ['AE 694', 'AES801', 'AES802', 'BB 701', 'BBS801', 'CE 694', 'CES801', 'CLS801', 'CMS801', 'CMS802', 'CS 694', 'CSS801', 'CSS802', 'DHS801', 'DSS801', 'ECS801', 'ECS802', 'EE 694', 'EES801', 'EN 406', 'EN 694', 'ENS801', 'ES 296', 'ES 694', 'ESS801', 'ET 694', 'GNRS801', 'GS 694', 'GSS801', 'HSS801', 'DE 687', 'DEP403', 'DEP706', 'IES801', 'LE 301', 'MA 450', 'MAS801', 'ME 694', 'MES801', 'MGS801', 'MGS802', 'MM 396', 'MM 433', 'MM 694', 'MMS801', 'PH 303', 'PHS801', 'PSS801', 'PSS803', 'SC 694', 'SCS801', 'SCS802', 'TD 694', 'TDE611', 'BB 694', 'BB 802', 'CHS802', 'CL 694', 'CLS802', 'CS 396', 'DS 694', 'EN 422', 'ENS 802', 'ES 312', 'ESS802', 'ETS801', 'GNR694', 'GNRS02', 'GSS802', 'HS 694', 'HSS802', 'IE 694', 'IES802', 'IES601', 'MMM694', 'PHS802', 'PS 694', 'PSS802', 'US 691', 'USS801']
seminar_df=df[df['Course Code'].isin(Seminar_courses)]
seminar_df.head(3)
```

```python
# Grouping by 'Year', and 'Course Code', taking sum for all columns except 'Year' and 'Course Code'
seminar_df = seminar_df.groupby(['Year', 'Course Code'], as_index=False).agg(
    {col: 'sum' for col in seminar_df.columns if col not in ['Year', 'Course Code']} |
    {'Year': 'first', 'Course Code': 'first'}
)
```

```python
seminar_df['Average Grade'] = (seminar_df['AP'] * 10 + seminar_df['AA'] * 10 + seminar_df['AB'] * 9 + seminar_df['BB'] * 8 + seminar_df['BC'] * 7 + seminar_df['CC'] * 6 + seminar_df['CD'] * 5 + seminar_df['DD'] * 4) / seminar_df['Total']
seminar_2020=seminar_df[seminar_df['Year']==2020]
seminar_2021=seminar_df[seminar_df['Year']==2021]
seminar_2022=seminar_df[seminar_df['Year']==2022]
seminar_2023=seminar_df[seminar_df['Year']==2023]
seminar_2020.sample(3)
```

```python
seminar_2020_export=pd.DataFrame()
seminar_2020_export['Course Code']=list(seminar_2020['Course Code'])
seminar_2020_export['Average Grade']=list(seminar_2020['Average Grade'])
seminar_2020_export.head()
```

```python
json_2020=seminar_2020_export.to_json(orient='records')

with open('records_2020.json', 'w') as json_file:
    json_file.write(json_2020)
```
